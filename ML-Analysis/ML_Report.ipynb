```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
%matplotlib inline
```

```python
def display_image(file_path):
    fig = plt.figure(figsize=(12, 10))
    img = mpimg.imread(file_path)
    imgplot = plt.imshow(img)
    plt.axis('off')
    plt.show()
    plt.close()
```

## 1 - Cluster Analysis

In this section, we experimented differents clusters on differents subjects to have an overview of the potential applications we can have on this dataset. Additionally, we aimed to identify what kind of cluster could be used for commercials and analysis purposes. 

### 1-1 General Cluster

For the first cluster, we just did a clusters analysis of diffente features (`city`, `customer_type`, `quantity` and `total_amount`) to observe how all customers would be grouped

```python
display_image("Plot_result/customer_depending_city_amount_type_features.png")
```

As we can see, the `customer_type` axis is clearly divided in 2 parts : `B2C` and `B2B`. So one cluster will represent the `B2C` customers and another the `B2B` customers.

Within the `B2B` customers, there are 4 differents clusters. These clusters are divided based on two criteria : the `total_amount` value (little and bigger) and the `city` value.
