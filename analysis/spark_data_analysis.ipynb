```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pandas as pd
%matplotlib inline
```

```python
def display_image(file_path):
    img = mpimg.imread(file_path)
    imgplot = plt.imshow(img)
    plt.axis('off')
    plt.show()
    plt.close()
```

## Purchases per month

In order to identify trends in product purchases,
we can study the products purchased for each month of the year.

### Main categories

```python
display_image("spark_analysis/mean_purchases_per_month.png")
```

We can identify three main trends in the kind of products purchased:
through most of the year,
the products with the biggest amount of purchases are books.
The amount of books bought gets slightly lower in June before rising again in August and September.
Furthermore, the amount of electronics purchased skyrockets in November and December,
right before the winter holidays.
We can also notice a slight increase in the purchases of products connected to sports and outdoor activities during the summer months.

### Sub categories

In order to get more details about these trends,
let's take a closer look at each sub-category.

#### Sports and Outdoors

```python
display_image("spark_analysis/mean_Sports_&_Outdoors_purchases_per_month.png")
```

The purchases of equipement for sport and outdoor activities do increase during the summer months,
but that is due to more people getting ready for camping during this period.

#### Books

```python
display_image("spark_analysis/mean_Books_purchases_per_month.png")
```

While there is no easily noticeable reason for the decrease of book purchases in June,
we can see that its increase in August and September is due to more people buying educational books.
This can be explained by the dates where school years begin, since they overlap.
